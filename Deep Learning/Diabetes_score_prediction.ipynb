```python
# 패키지 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 하이퍼 패러미터 설정
MY_BATCH = 64
MY_EPOCH = 300
```

```python
# 0. 데이터 불러오기
X, Y = load_diabetes(return_X_y=True, as_frame=True)
```

```python
# 1. 입력 데이터 시각화
X.boxplot()
plt.show()
```

```python
# 2. 출력 데이터 스케일링
scaler = StandardScaler()
Y = scaler.fit_transform(Y.values.reshape(-1, 1))
```

```python
temp = pd.DataFrame(Y)
temp.describe()
```

```python
# 3. 학습용, 평가용 데이터 분리
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
# 4. 인공신경망 구축
model = Sequential()

model.add(Dense(200, activation="relu", input_dim=10))
model.add(Dense(1000, activation="relu"))

model.add(Dense(1, activation="linear"))

model.summary()
```

```python
# 5. 기타 환경 설정
model.compile(optimizer="sgd", loss="mse", metrics=["acc"])
```

```python
# 6. 학습 실행 및 학습 내역 저장
hist = model.fit(X_train, Y_train, batch_size=MY_BATCH, epochs=MY_EPOCH, verbose=1, validation_data=(X_test, Y_test))
```

```python
# 7. 학습 내역 시각화
plt.plot(hist.history["loss"], label="train_loss")
plt.plot(hist.history["val_loss"], label="val_loss")

plt.legend()

plt.show()
```

```python
# 8. 평가 진행
loss, _ = model.evaluate(X_test, Y_test)
print(f"{loss=}")
```

```python
# 9. 예측 실행
Y_pred = model.predict(X_test)
```

```python
# 10. 예측 결과 시각화
plt.plot([-1.5, 2.5], [-1.5, 2.5], "r")
plt.plot(Y_test, Y_pred, "b.")
plt.show()
```

```python
# 11. 정규화 되돌리기
inversed_test = scaler.inverse_transform(Y_test)
inversed_pred = scaler.inverse_transform(Y_pred)
```

```python
# 12. 예측 결과 시각화
plt.plot([25, 350], [25, 350], "r")
plt.plot(inversed_test, inversed_pred, "b.")
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test, Y_pred)
r2
```
